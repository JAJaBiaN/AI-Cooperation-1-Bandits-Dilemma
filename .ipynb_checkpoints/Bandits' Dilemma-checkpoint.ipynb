```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
class Random_bandit:
    
    def __init__(self):
        self.needs_update = False
        return None
    
    def select(self):
        """Randomly choose one of two options"""
        if np.random.rand()<0.5:
            return 0
        else:
            return 1
    
    def reset(self):
        """Pretend to reset as if creating a new agent
            Note: The agent doesn't learn so resetting is pointless"""
        return None
```

```python
class Rational_bandit:
    
    def __init__(self):
        self.needs_update = False
        return None
    
    def select(self):
        """Selects the optimal choice based on game theory (This is always 1 because of the setup)"""
        return 1
    
    def reset(self):
        """Pretend to reset as if creating a new agent
            Note: The agent doesn't learn so resetting is pointless"""
        return None
```

```python
class Optimal_bandit:
    def __init__(self):
        self.been_betrayed = False
        
        self.needs_update = True
        return None
    
    def select(self):
        if self.been_betrayed:
            return 1
        else:
            return 0
    
    def update(self, choice, r, t):
        """Find out if the bandit has been betrayed this round
            Note: the default payoff matrix has been hard-coded in for simplicity"""
        if not self.been_betrayed:
            if r == -3:
                self.been_betrayed = True
        return self.been_betrayed
    
    def reset(self):
        """Reset as if creating a new agent"""
        self.been_betrayed = False
        return None
```

```python
class eGreedy_bandit:
    def __init__(self, k, e, alpha=None):
        self.k = k
        self.e = e
        self.alpha = alpha
        self.Q = np.zeros(k)
        
        self.needs_update = True
        return None
    
    def select(self):
        if np.random.rand()>self.e:
            return np.argmax(self.Q)
        else:
            if np.random.rand()<0.5:
                return 0
            else:
                return 1
    
    def update(self, choice, r, t):
        if self.alpha == None:
            self.Q[choice] = self.Q[choice] + (r - self.Q[choice]) / t
        else:
            self.Q[choice] = self.Q[choice] + self.alpha * (r - self.Q[choice])
        return self.Q
    
    def reset(self):
        """Reset as if creating a new agent"""
        self.Q = np.zeros(self.k)
        return None
```

```python
class UCB_bandit:
    def __init__(self, k, c, alpha=None):
        self.k = k
        self.c = c
        self.alpha = alpha
        self.count_chosen = np.zeros(k)
        self.Q = np.zeros(k)
        
        self.needs_update = True
        return None
    
    def select(self, t):
        return np.argmax(self.Q + self.c * np.sqrt(np.log(t) / self.count_chosen))
    
    def update(self, choice, r, t):
        # Update the number of times the choice has been made
        self.count_chosen[choice] += 1
        
        # Update Q
        if self.alpha == None:
            self.Q[choice] = self.Q[choice] + (r - self.Q[choice]) / t
        else:
            self.Q[choice] = self.Q[choice] + self.alpha * (r - self.Q[choice])
        return self.Q
    
    def reset(self):
        """Reset as if creating a new agent"""
        self.count_chosen = np.zeros(self.k)
        self.Q = np.zeros(self.k)
        return None
```

```python
class No_Regret_bandit:
    def __init__(self, k, largest_cost):
        """A bandit that uses a no-regret algorithm (multiplicative weights) to select a mixed strategy and then an action."""
        self.k = k
        self.largest_cost = largest_cost
        self.approx_T = 2. #using lowest power of two > t to account for unknown number of rounds
        self.eps = np.sqrt(np.log(self.k) / self.approx_T)
        self.weights = np.ones(k)
        
        self.needs_update = True
        return None
    
    def select(self):
        Gamma = np.sum(self.weights)
        random_num = np.random.rand()
        for i in range(self.k):
            if random_num < np.sum(self.weights[0:i+1]) / Gamma:
                # Since np.random.rand() produces a result in [0,1), one option will be chosen
                return i
                         
    def update(self, choice, r, t):
        # Update eps for the approximate T
        if t > self.approx_T:
            self.approx_T *= 2. #use the next power of 2
            self.eps = np.sqrt(np.log(self.k) / self.approx_T)
        
        # Determine the cost vector, c, based on choice and reward
        #### NOTE: This is hard-coded based on the default reward matrix, it will not work for others!
        if r == -1 or r == 0:
            c = np.array([1, 0])
        elif r == -2 or r == -3:
            c = np.array([3, 2])
        
        # Scale the reward so that the largest possible cost is 1
        c = c / self.largest_cost
        
        # Update the weights
        self.weights = self.weights * (1 - self.eps)**c
        return self.weights
    
    def reset(self):
        """Reset as if creating a new agent"""
        self.approx_T = 2 #using lowest power of two > t to account for unknown number of rounds
        self.eps = np.sqrt(np.log(self.k) / self.approx_T)
        self.weights = np.ones(self.k)
        return None
```

```python
def share_rewards(payoff_matrix, share):
    """Force empathy with the other bandit by taking a weighted average of rewards"""
    payoff_shape = payoff_matrix.shape
    
    new_payoffs = []
    for R in payoff_matrix.reshape(-1,2):
        r0 = share * R[1] + (1-share) * R[0] 
        r1 = share * R[0] + (1-share) * R[1]
        new_payoffs.append((r0, r1))
    
    new_payoffs = np.array(new_payoffs).reshape(payoff_shape)
    return new_payoffs

def reward(choice_1, choice_2, payoff_matrix):
    return payoff_matrix[choice_1, choice_2]
```

```python
def run_expt(T, b_type0, b_type1, filename_addition, num_runs=100, e1=0.15, e2=None, c1=1, c2=None, learn_over_episodes = False, payoff_matrix = np.array([[[-1, -1], [-3, 0]], [[0, -3], [-2,-2]]])):
    """Run a prisoner's dilemma experiment for two bandits with two choices"""
    
    # Create a filename to store the output in
    save_filename = filename_addition
    
    # Initialise the bandit
    if b_type0 == "eGreedy":
        bandit0 = eGreedy_bandit(payoff_matrix.shape[0], e1)
        save_filename += "eGreedy bandit e="+str(e1)
    elif b_type0 == "UCB":
        bandit0 = UCB_bandit(payoff_matrix.shape[0], c1)
        save_filename += "UCB bandit c="+str(c1)
    elif b_type0 == "No-regret":
        bandit0 = No_Regret_bandit(payoff_matrix.shape[0], 3)
        save_filename += "No-regret bandit"
    elif b_type0 == "Rational":
        bandit0 = Rational_bandit()
        save_filename += "Rational bandit"
    elif b_type0 == "Optimal":
        bandit0 = Optimal_bandit()
        save_filename += "Optimal bandit"
    elif b_type0 == "Random":
        bandit0 = Random_bandit()
        save_filename += "Random bandit"
    
    if b_type1 == "eGreedy":
        if e2 == None:
            e2 = e1
        bandit1 = eGreedy_bandit(payoff_matrix.shape[0], e2)
        save_filename += " vs eGreedy bandit e="+str(e2)
    elif b_type1 == "UCB":
        if c2 == None:
            c2 = c1
        bandit1 = UCB_bandit(payoff_matrix.shape[0], c2)
        save_filename += " vs UCB bandit c="+str(c2)
    elif b_type1 == "No-regret":
        bandit1 = No_Regret_bandit(payoff_matrix.shape[0], 3)
        save_filename += " vs No-regret bandit"
    elif b_type1 == "Rational":
        bandit1 = Rational_bandit()
        save_filename += " vs Rational bandit"
    elif b_type1 == "Optimal":
        bandit1 = Optimal_bandit()
        save_filename += " vs Optimal bandit"
    elif b_type1 == "Random":
        bandit1 = Random_bandit()
        save_filename += " vs Random bandit"
    
    save_filename += ", T="+str(T)
    save_filename += ", "+str(num_runs)+" episodes"
    
    # Initialise storage
    single_rewards0 = []
    single_rewards1 = []
    choices0 = []
    choices1 = []
    
    # Counters
    for run in range(num_runs):
        # Print update
#         print("Run:", run)
        
        # Check whether we want agents to remember previous episodes
        if not learn_over_episodes:
            # Reset bandits
            bandit0.reset()
            bandit1.reset()
    
        single_rewards0.append([])
        single_rewards1.append([])
        choices0.append([])
        choices1.append([])
        
        # Run experiment for T rounds
        for t in range(1, T+1): #rounds

            # Update rewards
            if b_type0 == "UCB":
                choice0 = bandit0.select(t)
            else:
                choice0 = bandit0.select()
            
            if b_type1 == "UCB":
                choice1 = bandit1.select(t)
            else:
                choice1 = bandit1.select()
            
            r0, r1 = reward(choice0, choice1, payoff_matrix)
            
            if bandit0.needs_update:
                bandit0.update(choice0, r0, t)
            if bandit1.needs_update:
                bandit1.update(choice1, r1, t)
            
            single_rewards0[run].append(r0)
            single_rewards1[run].append(r1)
            choices0[run].append(choice0)
            choices1[run].append(choice1)
    
    # Work out title
    if b_type0 == b_type1:
        if b_type0 == "eGreedy":
            title_string = b_type0 + " Bandit's Dilemma, e1="+str(e1)+", e2="+str(e2)
        elif b_type0 == "UCB":
            title_string = b_type0 + " Bandit's Dilemma, c1="+str(c1)+", c2="+str(c2)
        else:
            title_string = b_type0 + " Bandit's Dilemma"
    else:
        title_string = b_type0 + " v " + b_type1 + " Bandit's Dilemma"
        
        if b_type0 == "eGreedy":
            title_string = title_string +", e1="+str(e1)
        elif b_type0 == "UCB":
            title_string = title_string +", c1="+str(c1)
        
        if b_type1 == "eGreedy":
            title_string = title_string +", e2="+str(e2)
        elif b_type1 == "UCB":
            title_string = title_string +", c2="+str(c2)
    
    #################################
    ## Save data and some metadata ##
    #################################
    # Make the folder
    os.mkdir(save_filename)
    
    # Write metadata to a file
    f = open("./"+save_filename+"/metadata.txt", "w+")
    
    #   Information about bandit 0
    f.write("Bandit 0 learning type: "+b_type0)
    if b_type0 == "eGreedy":
        f.write("\n\te_0 is set as: "+str(e1))
    elif b_type0 == "UCB":
        f.write("\n\tc_0 is set as: "+str(c1))
    
    #   Information about bandit 1
    f.write("\nBandit 1 learning type: "+b_type1)
    if b_type1 == "eGreedy":
        f.write("\n\te_1 is set as: "+str(e2))
    elif b_type1 == "UCB":
        f.write("\n\tc_1 is set as: "+str(c2))
    
    #   Information on experimental set up
    f.write("\n\nNumber of timesteps per run: "+str(str(T)))
    f.write("\nNumber of runs: "+str(num_runs))
    
    #   Payoff matrix
    f.write("\n\nPayoff matrix:")
    f.write("\n\t"+str(payoff_matrix[0, 0])+" | "+str(payoff_matrix[0, 1]))
    f.write("\n\t"+str(payoff_matrix[1, 0])+" | "+str(payoff_matrix[1, 1]))
    
    f.close()
    
    # Save arrays as a csv file
    df = pd.DataFrame({"Run 0, Choices 0":choices0[0], "Run 0, Choices 1":choices1[0], "Run 0, Rewards 0":single_rewards0[0], "Run 0, Rewards 1":single_rewards1[0]})
    for run in range(1, num_runs):
        df["Run "+str(run)+", Choices 0"] = choices0[run]
        df["Run "+str(run)+", Choices 1"] = choices1[run]
        df["Run "+str(run)+", Rewards 0"] = single_rewards0[run]
        df["Run "+str(run)+", Rewards 1"] = single_rewards1[run]
    df.to_csv("./"+save_filename+"/array data.csv")
    
    return choices0, choices1, single_rewards0, single_rewards1
```
